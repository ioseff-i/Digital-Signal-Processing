# Linear and Cubic Interpolation

```python
from scipy.interpolate import interp1d

x = np.linspace(0,4,120)
y = np.cos(x**2/3+4)
plt.plot(x,y,'o')
```

```python
f1 = interp1d(x,y,kind='linear')
f2 = interp1d(x,y,kind='cubic')

plt.plot(x,y,'o',x,f1(x),'-',x,f2(x),'--')
```

# Waveform Generator

## Gausspulse

```python
t = np.linspace(-1,1,2*100,endpoint=False)
i,q,e = signal.gausspulse(t,fc=5,retquad=True,retenv=True)

fig,axs = plt.subplots(3,1)
plt.sca(axs[0])
plt.plot(t,i)

plt.sca(axs[1])
plt.plot(t,q)

plt.sca(axs[2])
plt.plot(t,e)

plt.show()
```

```python
plt.plot(t,i,t,q,t,e,'--')
```

## Square Waves,pulse-width modulated sine waves

```python
t= np.linspace(0,1,500,endpoint=False)
square_wave = signal.square(2*np.pi*5*t)
plt.plot(square_wave)
```

```python
sine_sig = np.sin(2*np.pi*t)
pwm = signal.square(2*np.pi*30*t,duty=(sine_sig)/2)
fig,axs = plt.subplots(2,1)
plt.sca(axs[0])
plt.plot(t,sine_sig)

plt.sca(axs[1])
plt.plot(t,pwm)


plt.show()
```
