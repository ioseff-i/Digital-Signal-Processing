# Basic Filter example 

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
from scipy.fftpack import fft,ifft
import mysignals as sigs
%matplotlib inline
style.use('ggplot')
plt.rcParams['figure.figsize'] = [10,8]
plt.rcParams.update({'font.size':18})
```

```python
signal = sigs.InputSignal_1kHz_15kHz
```

```python
def filter(x,cutoff):
    N = len(x)
    output = [0]*N
    for i in range(N):
        if(x[i]>cutoff):
            output[i]=x[i]
    return output
```

```python
output=filter(signal,-0.1)
```

```python
fig,axs = plt.subplots(2,1)

plt.sca(axs[0])
plt.plot(signal,color='r',LineWidth=1.5,label='Signal Itself')
plt.legend()

plt.sca(axs[1])
plt.plot(output,color='b',LineWidth=1.5,label='Filtered')
plt.legend()

plt.legend()

plt.show()
```

```python
n = np.linspace(0,5,10000)
signal = np.cos(2*np.pi*n)+np.sin(4*np.pi*n)


def window(x):
    N = len(x)
    output = [None]*len(x)
    for n in range(len(x)):
        output[n] = 0.42- 0.5*np.cos(2*np.pi*x[n]/(N-1))+0.08*np.cos(4*np.pi*x[n]/(N-1))
    return output
```

```python
fig,axs = plt.subplots(2,1)

plt.sca(axs[0])
plt.plot(signal,color='r',LineWidth=1.5,label='Signal Itself')
plt.legend()

plt.sca(axs[1])
plt.plot(window(signal),color='b',LineWidth=1.5,label='Low pass filter with OpenCVFiltered')
plt.legend()

plt.legend()

plt.show()
```
