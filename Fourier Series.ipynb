# Fourier Series

```python
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import style
style.use('ggplot')
plt.rcParams['figure.figsize'] = [10,8]
plt.rcParams.update({'font.size':18})
```

## Define the square wave function

```python
T = 3
dt = 0.0001
tmax = 15
n_max = 1

N = int(tmax/dt)
t = np.arange(0,N)*dt
y = np.zeros(N)

y[np.floor(2*t/T)%2==0] = 1

plt.figure(figsize=(15,6))
plt.plot(t,y,color='green')
plt.ylim((-.2,1.2))
plt.xlim(0,tmax)
plt.xlabel('time (s)')
```

## Fourier Series of the Square wave function    
    

```python
n_max = 30
n = np.arange(1,n_max+1)
a = list()
b = list()
y_simul = np.zeros(N) + 1/2
for i in range(n_max):
    a.append(0)
    if i%2==0:
        b.append(0)
    else:
        b.append(2/(np.pi*i))
    y_simul = y_simul + a[-1]*np.cos(2*np.pi*i*t/T) + b[-1]*np.sin(2*np.pi*i*t/T)

plt.figure(figsize=(15,6))
plt.plot(t,y_simul,color='orange')
plt.title('n = '+str(n))
```

## So the together:


```python
plt.figure(figsize=(15,6))
plt.plot(t,y,color='green')
plt.ylim((-.2,1.2))
plt.xlim(0,tmax)
plt.xlabel('time (s)')
plt.plot(t,y_simul,color='orange')
plt.title('n = '+str(n))
plt.savefig('square30.png')
```

## Half Ramp Signal 

```python
T = 2
dt = 0.001
tmax = 15
n_max = 10

N = int(tmax/dt)
t = np.arange(0,N)*dt
y = np.zeros(N)

ii = ((2*t/T) % 2 < 1)
y[ii]=t[ii]-np.floor(t[ii])

plt.figure(figsize=(15,6))
plt.plot(t,y,color = 'green')
plt.ylim((-.2,1.2))
plt.xlim(0,tmax)
```

## Fourier Series of the half ramp function

```python
n = np.arange(1,n_max+1)
a = list()
b = list()
y_simul = np.zeros(N) + T/8
for i in n:
    if i % 2 == 0:
        a.append(0)
        b.append(-T/(2*np.pi*i))
    else:
        a.append(-T/(np.pi*np.pi*i*i))
        b.append(T/(2*np.pi*i))
    y_simul = y_simul + a[-1]*np.cos(2*np.pi*i*t/T) + b[-1]*np.sin(2*np.pi*i*t/T) 

plt.figure(figsize=(15,6))
plt.plot(t,y_simul,color='orange')
plt.title('n = '+str(n))
plt.grid()
```

## Putting together

```python
plt.figure(figsize=(15,6))
plt.plot(t,y,color = 'green')
plt.ylim((-.2,1.2))
plt.xlim(0,tmax)
plt.plot(t,y_simul,color='orange')
plt.title('n = '+str(n))
plt.grid()
```
