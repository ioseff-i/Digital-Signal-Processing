# Transfer Function 

```python
from scipy import signal
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
from scipy.fftpack import fft,ifft
import mysignals as sigs
%matplotlib inline
style.use('ggplot')
plt.rcParams['figure.figsize'] = [10,8]
plt.rcParams.update({'font.size':18})
```

```python
#H(s) = s^2 + 3s + 3/s^2+2s

num = [1,3,3]
denom = [1,2,1]

H1 = signal.TransferFunction(num,denom)

H2 = signal.TransferFunction(num,denom,dt=0.1)

print(H1)
print(H2)

```

# Zeros Poles Gain

```python
# H(s) = 5(s-1)(s-2)/(s-3)(s-4)
simple_lti = signal.ZerosPolesGain([1,2],[3,4],5)
simple_dlti = signal.ZerosPolesGain([1,2],[3,4],5,dt = 0.1)
print(simple_lti)
print(simple_dlti)
```

# Frequency Response of LTI

```python
s1 = signal.lti([],[1,1,1],[5])
w,H = signal.freqresp(s1)
#Complex plot
plt.plot(H.real,H.imag)
```

# Bode Plot

```python
# H(s) = s^2/s^2+s
s1 = signal.lti([1],[1,1])
w,mag,phase = s1.bode()
plt.semilogx(w,mag)
plt.figure()
plt.semilogx(w,phase)
plt.show()
```

# Impulse Response of LTI

```python
# H(s) = s^2 /s^3+2s^2+s
system = ([1],[1,2,1])
t,y = signal.impulse(system)
plt.plot(t,y)
```
